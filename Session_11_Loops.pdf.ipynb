```python
# Case-1: pwd correct code should stop we did

# Internet login probelm
# when I enter a password
#   If it is correct : successfully logged in
#   If it is fail   : wrong password try one more time
#                     you have 3 chances left


# suppose you attempted 3 times wrong password
# all chances are gone
# print try again after 24hours
```

```python
correct_pwd = "54678"     
attempts = 3                  

while attempts > 0:
    pwd = input("Enter your password: ")
    
    if pwd == correct_pwd:
        print("Successfully Logged In!")
        break   
    else:
        attempts = 1
        if attempts > 0:
            print(f"Wrong Password! Try again. You have {attempts} chances left.")
        else:
            print("All chances gone! Try again after 24 hours.")
```
