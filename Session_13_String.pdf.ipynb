```python
def extract(str1):
    i1=str1.find('.')  
    i2=str1.find('@') 
    i3=str1.find('.',i1+1)
    i1,i2,i3
    fname=str1[:i1]
    sname=str1[i1+1:i2]
    cname=str1[i2+1:i3]
    return(fname,sname,cname)
extract('Shreyansh.tiwari@cognizant.com')
```
