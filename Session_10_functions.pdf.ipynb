- function is returning the outputs to me

- i need to store using **return** keyword

```python
def math_fun1():
    n1=eval(input('Enter a number:'))
    n2=eval(input('Enter a number:'))
    Add=n1+n2   # local
    Sub=n1-n2   # local
    print(f'the sum of {n1},{n2} are:{Add}')
    print(f'the sub of {n1},{n2} are:{Sub}')
    return(Add,Sub)
```

```python
def sum_n(n):
    if n == 1:
        return 1
    else:
        return n + sum_n(n - 1)

print(sum_n(5))  
```

```python
sum_n(5)
= 5 + sum_n(4)
= 5 + (4 + sum_n(3))
= 5 + (4 + (3 + sum_n(2)))
= 5 + (4 + (3 + (2 + sum_n(1))))
= 5 + (4 + (3 + (2 + 1)))
= (15)
```

```python
def factorial(n):
    if n == 1:             
        return 1
    else:                  
        return n * factorial(n - 1)

print(factorial(5))
```

```python
factorial(5)
= 5 * factorial(4)
= 5 * (4 * factorial(3))
= 5 * (4 * (3 * factorial(2)))
= 5 * (4 * (3 * (2 * factorial(1))))
= 5 * 4 * 3 * 2 * 1 = 120
```
