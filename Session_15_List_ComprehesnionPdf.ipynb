$Case-3:for -if-else$

```python
for i in Input:
    if '#' in in i:
        print(i)
    else:
        print(i)
```

```python
# [<output>  <for loop>}
# [<output>  <for loop>  <if condition>] 
# [<if_out> <if_con>  else  <else out> <for loop>]
```

```python
Input=[20,21,20,31,34]
out=[]
for i in Input:
    if i%2==0:
        out.append(f'even_{i}')
    else:
        out.append(f'odd_{i}')
out
```

```python
# if_out:f'even_(i)'
# if_con: if i%2==0
# else_out: f'odd_(i)'
# for i in Input

[f'even_{i}' if i%2==0 else f'odd_{i}' for i in Input]
```

$Case-4:if-elif-else$

```python
[<output>  <for loop>}
[<output>  <for loop>  <if condition>] 
[<if_out> <if_con>  else  <else out> <for loop>]
[<if_out> <if_con> else <elif out> <elif cond> else <else out> <for loop>]


# Rule-1: there is no elif
         # else if
# Rule-2: if after else
# Rule-3: else after output

Input=[50,70,90]   # C B A
out=[]
for i in Input:
    if i>==90:
        out.append('A')
    elif i>=60:
        out.append('B')
    else:
        out.append('c')
```

```python
# if_out if_cond \
# else elif_out elif_cond \
# else elif_out elif_cond \
# else elif_out elif_cond \
# else else_out
Input=[50,70,90]

['A' if i>=90 \
else 'B' if i>=60 \
else 'C' \
for i in Input]
```

```python
# [<out> <for loop>]
# <out> <for> <if>
# [<if out> <if cond> else <els out> <for loop>]
# <if out> <if> else <elif out> <elif con> else <else out> <>
```
