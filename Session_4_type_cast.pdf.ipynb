# Type cast

- convert one data type to another data type

```python
a=100
b=200
c=a+b
c
```

```python
a='100' 
b='200'
c=a+b
c
```

- in above case the both string are concatenated

- if we wants to get a correct answer we need convert string to integer

- this is the one example of **type cast**
