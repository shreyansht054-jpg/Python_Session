```python
l1=[10,20,30,40]
l2=[1,2,3]
#l=[11,22,33,40]
l=[]
for i in range(len(l1)):  # 0,1,2,3
    if i<len(l2):
        l.append(l1[i]+l2[i])
    else:
        l.append(l1[i])
# step-1: i=0  if 0<3  True  0 10+1=11
# step-2: i=1  if 1<3  True  1 20+2=22
# step-3: i=2  if 2<3  True  2 30+3 =33
# step-4: i=3  if 3<3  F  else 3 40
```

**Dictionary Comprehension**

```python
for i,j in zip(names,age):
    d[i]=j
[d[i]=j for i,j in zip(names,age)]

for i in numbers:
    if i%2==0:
        d[i]='even'
    else:
        d[i]='odd'

[d[i]='even' if i%2==0 else d[i]='odd' for i in numbers]
```
