**Finally block**

- **try** block is used to run original code or correct code

- **except** block is used to catch the error mainly  

- in colud whenever we establish a connection money deducton start

- even for correct code or wrong code the money deducton start when connention establish

- so it is very important we need to close the connection

- in that sitution we will use **finally** block

- if try block executes entirely after finally block also execute

- if except block executes entirely after finally block also execute

```python
try:
    a=10
    b=20
    c=a+b
    print(f"the addition of {a} and {b} is: {c}")

except:
    print('Hello bro error agaya thats im agaya')
    
finally:
    print('bye')
```
