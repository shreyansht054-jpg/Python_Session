```python
 # wap ask the user take two number print subtraction multiplication and division

a = eval(input("Enter first number: "))
b = eval(input("Enter second number: "))

print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
# wap ask the user take the radius find the area of circle
    # pi=3.14
    # r=100
   # area_of_circle=pi*r*r

pi = 3.14
r = eval(input("Enter the radius of the circle: "))

area_of_circle = pi * r * r
print("Area of Circle:", area_of_circle)
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h

h = eval(input("Enter height of the triangle: "))
b = eval(input("Enter breadth of the triangle: "))

area_of_triangle = 0.5 * b * h
print("Area of Triangle:", area_of_triangle)
```

```python
# wap ask the user take 3 numbers find the average

x = eval(input("Enter first number: "))
y = eval(input("Enter second number: "))
z = eval(input("Enter third number: "))

average = (x + y + z) / 3
print("Average:", average)
```

```python
# wap ask the user take the bill amount and tip amount calculate total bill

bill = eval(input("Enter bill amount: "))
tip = eval(input("Enter tip amount: "))

total = bill + tip
print("Total Bill:", total)
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount

salary = eval(input("Enter your salary: "))
tax_percent = eval(input("Enter tax percentage: "))

tax_amount = (tax_percent / 100) * salary
print("Tax Amount:", tax_amount)
```

```python
# wap ask the user take the value in dollar convert into rupees
#         1$=80rs

dollars = eval(input("Enter amount in dollars: "))
rupees = dollars * 80

print("Amount in Rupees:", rupees)
```

```python
# wap ask the use take weight in kgs convert into pounds
#     1kg=2.2pounds

kgs = eval(input("Enter weight in kilograms: "))
pounds = kgs * 2.2

print("Weight in Pounds:", pounds)
```
